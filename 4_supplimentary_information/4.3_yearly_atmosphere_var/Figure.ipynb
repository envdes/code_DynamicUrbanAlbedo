# Atmosphere variables figure
- This script is used to visualize atmosphere variables from 2015 to 2099;

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

```python
f0 = 'atm.csv'
df0 = pd.read_csv(f0)
for column_name in df0.columns:
    globals()[column_name] = df0[column_name]
```

```python
# drawing
linewidth = 0.5
linewidth2 = 1
fontsize = 6
start_year = 2015
end_year = 2100
labelcolor = '#6b6b6b'
lc1 = '#6e1913'
lc2 = '#a42018'
lc3 = '#f62d18'
lc4 = '#d97c26'
lc5 = '#ff912b'
lc6 = '#30b5d6'
lc7 = '#28ac92'
lc8 = '#123264'
lc9 = '#05101f'
lc10 = '#952cc6'
lc11 = '#2c4dc6'
lc12 = '#5fc62c'
minor_locator = MultipleLocator(10)
specific_years = [2015, 2050, 2099]
alpha=0.075
xwidth = 0.75
length = 4
labelfont = 6
linestyle='dashed'
pad=1
handletextpad=0.5
padding = 4
data = [[Arr_TBOT_1, Arr_TBOT_1_JJA, Arr_TBOT_1_DJF], 
        [Arr_FSDS_1, Arr_FSDS_1_JJA, Arr_FSDS_1_DJF],
        [Arr_FLDS_1, Arr_FLDS_1_JJA, Arr_FLDS_1_DJF],
        [Arr_QBOT_1, Arr_QBOT_1_JJA, Arr_QBOT_1_DJF],
        [Arr_PBOT_1, Arr_PBOT_1_JJA, Arr_PBOT_1_DJF],
        [Arr_RAIN_1, Arr_RAIN_1_JJA, Arr_RAIN_1_DJF],
        [Arr_SNOW_1, Arr_SNOW_1_JJA, Arr_SNOW_1_DJF]]
l1 = ['ANN', 'JJA', 'DJF']
l2 = ['Near-term', 'Mid-term', 'Long-term']
yminlist=[10, 140, 300, 0.006, 95200/100, 0.00002*1000, -0.000001*1000]
ymaxlist=[30, 240, 425, 0.016, 95900/100, 0.00005*1000, 0.000004*1000]
t = ['TBOT', 'FSDS', 'FLDS', 'QBOT', 'PBOT', 'RAIN', 'SNOW']
yl = ['℃', 'W m⁻²', 'W m⁻²', 'kg kg⁻¹', 'hPa', 'mm ms⁻¹', 'mm ms⁻¹']
lc = [lc1, lc4, lc7, lc9, lc5, lc2]
```

```python
fig, axes = plt.subplots(nrows=2, ncols=4, figsize=(7.4, 2.5))
plt.subplots_adjust(left=0.035, right=0.98, bottom=0.06, top=0.94, wspace=0.25, hspace=0.5)
ax = axes.flatten()
for i in range(7):
    for j in range(3):
        if i==0 :
            ax[i].plot(time,data[i][j], linewidth=linewidth, label=l1[j], color = lc[j])
        elif i==4:
            ax[i].plot(time,data[i][j]/100, linewidth=linewidth, label=l1[j], color = lc[j])
        elif i>4:
            ax[i].plot(time,data[i][j]*1000, linewidth=linewidth, label=l1[j], color = lc[j])
        else:
            ax[i].plot(time,data[i][j], linewidth=linewidth, color = lc[j])
    ax[i].set_ylim(yminlist[i], ymaxlist[i])        
    ax[i].fill_betweenx([yminlist[i], ymaxlist[i]],2021, 2040, color=lc9, alpha=alpha, label='Near-term',linewidth=0)
    ax[i].fill_betweenx([yminlist[i], ymaxlist[i]],2041, 2060, color=lc5, alpha=alpha, label='Mid-term',linewidth=0)
    ax[i].fill_betweenx([yminlist[i], ymaxlist[i]],2081, 2100, color=lc2, alpha=alpha, label='Long-term',linewidth=0)
    ax[i].set_title(t[i],fontsize=fontsize,color=labelcolor, pad = padding)
    ax[i].set_xlim(start_year,2099)
    ax[i].tick_params(axis='y', which='both', left=False, right=False, labelleft=True, labelcolor=labelcolor, color = labelcolor, labelsize=labelfont,pad=pad, width = xwidth, length = 0)
    ax[i].tick_params(axis='x', which='major', top=False, bottom=True, labelbottom=True, labelcolor=labelcolor, color = labelcolor, labelsize=labelfont,pad=pad, width = xwidth, length = length)
    ax[i].tick_params(axis='x', which='minor', top=False, bottom=True, labelbottom=True, labelcolor=labelcolor, color = labelcolor, labelsize=labelfont,pad=pad, width = xwidth, length = length/2)
    ax[i].spines['left'].set_visible(False)
    ax[i].spines['top'].set_visible(False)
    ax[i].spines['right'].set_visible(False)
    ax[i].spines['bottom'].set_color(labelcolor)
    ax[i].xaxis.set_minor_locator(minor_locator)
    ax[i].set_xticks(specific_years)
    ax[i].grid(True, which='major', axis='y',linestyle='--',linewidth=0.5, alpha=0.5)
    ax[i].set_ylabel(yl[i], labelpad=5, rotation=0, ha='center', fontsize=labelfont,color=labelcolor)
    ax[i].yaxis.set_label_coords(0, 1.05)

fig.delaxes(axes[1, 3])
lines, labels = axes[0, 0].get_legend_handles_labels()
legend = fig.legend(lines, labels,bbox_to_anchor=(0.75, 0.03), loc='lower left',frameon=False, fontsize=labelfont)
for i, text in enumerate(legend.get_texts()):
    text.set_color(lc[i])   
        
fig.savefig('atmospheric_forcing.png', dpi=300)
```
