# Regression
- This script is used to run fixed effect regression model;
- Simulations: CNTL, ROOF_TV, IMPROAD_TV, WALL_TV, ROOF_IMPROAD_TV, ROOF_IMPROAD_WALL_TV;

```python
from sklearn.metrics import r2_score
import statsmodels.api as sm
from linearmodels.panel import PanelOLS
import xarray as xr
import numpy as np
import pandas as pd
```

```python
# fix effect model
df = pd.read_csv('case2_export_annual.csv')
df['id'] = df['lsmlat'].astype(str) + '_' + df['lsmlon'].astype(str)
dff = df.set_index(['id', 'year'])
exog = sm.add_constant(dff[['alb', 'FSDS']])
grunfeld_fet = PanelOLS(dff['FSA_U'], exog, entity_effects=True, time_effects=False)
print(grunfeld_fet.fit())
```
