# Export urban temperature variables
- This script is used to export CUHI, SUHI, SWBGT, TBUILD variables;
- Simulations: CNTL, ROOF_TV, IMPRAOD_TV, WALL_TV;

```python
import os
import xarray as xr
import numpy as np
import pandas as pd
```

```python
home_path = '/work/n02/n02/yuansun/'
# surface
sfile = home_path + 'cesm/cesm_inputdata/lnd/clm2/surfdata_map/release-clm5.0.18/surfdata_0.9x1.25_hist_16pfts_Irrig_CMIP6_simyr1850_c190214.nc'
#case0
path_0 = home_path + 'cesm/archive/case0/lnd/hist/'
#case2
path_2 = home_path + 'cesm/archive/case2/lnd/hist/'
#case3
path_3 = home_path + 'cesm/archive/case3/lnd/hist/'
#case5
path_5 = home_path + 'cesm/archive/case5/lnd/hist/'
```

```python
start_year = 2015
end_year = 2100
ds_0 = xr.open_dataset(sfile)
mask = np.any(ds_0['PCT_URBAN'] != 0, axis=0)
mask = mask.rename({'lsmlat': 'lat', 'lsmlon': 'lon'})
```

```python
# case0 UHI
Arr_CUHI_0 = []
Arr_CUHI_0_JJA = []
Arr_CUHI_0_DJF = []
Arr_SUHI_0 = []
Arr_SUHI_0_JJA = []
Arr_SUHI_0_DJF = []
Arr_SWBGT_U_0 = []
Arr_SWBGT_U_0_JJA = []
Arr_SWBGT_U_0_DJF = []
Arr_TBUILD_0 = []
Arr_TBUILD_0_JJA = []
Arr_TBUILD_0_DJF = []
time = []

for year in range(start_year,end_year):
        fn = 'case0.clm2.h1.' + '%04.0f' % year + '-02-01-00000.nc'
        year_1 = year +1
        fn_1 = 'case0.clm2.h1.' + '%04.0f' % year_1 + '-02-01-00000.nc'
        if(os.path.exists(path_0+fn)):
            ds = xr.open_dataset(path_0+fn)   # 打开第一个文件
            ds_1 = xr.open_dataset(path_0+fn_1)
            time.append(year)
            
            cuhi = (ds['TSA_U']- ds['TSA_R']).where(mask)
            cuhi_1 = (ds_1['TSA_U']- ds_1['TSA_R']).where(mask)
            suhi = (ds['TG_U']- ds['TG_R']).where(mask)
            suhi_1 = (ds_1['TG_U']- ds_1['TG_R']).where(mask)
            SWBGT = ds['SWBGT_U']
            SWBGT_1 = ds_1['SWBGT_U']
            tbuild = ds['TBUILD'] -273.15
            tbuild_1 = ds_1['TBUILD'] -273.15
            
            Arr_CUHI_0.append(cuhi.mean().item())
            Arr_CUHI_0_JJA.append(cuhi[6:9,:,:].mean().item())
            Arr_CUHI_0_DJF.append(xr.concat([cuhi.isel(time=11),cuhi_1.isel(time=slice(0, 2))], 
                                 dim='time').mean().item())
            
            Arr_SUHI_0.append(suhi.mean().item())
            Arr_SUHI_0_JJA.append(suhi[6:9,:,:].mean().item())
            Arr_SUHI_0_DJF.append(xr.concat([suhi.isel(time=11), suhi_1.isel(time=slice(0, 2))], 
                                 dim='time').mean().item())
            
            Arr_SWBGT_U_0.append(SWBGT.mean().item())
            Arr_SWBGT_U_0_JJA.append(SWBGT[6:9,:,:].mean().item())
            Arr_SWBGT_U_0_DJF.append(xr.concat([SWBGT.isel(time=11), SWBGT_1.isel(time=slice(0, 2))], 
                                 dim='time').mean().item())
            
            Arr_TBUILD_0.append(tbuild.mean().item())
            Arr_TBUILD_0_JJA.append(tbuild[6:9,:,:].mean().item())
            Arr_TBUILD_0_DJF.append(xr.concat([tbuild.isel(time=11), tbuild_1.isel(time=slice(0, 2))], 
                                 dim='time').mean().item())
                                            
```

```python
data_dict0 = {
    'time': time,
    'Arr_CUHI_0': Arr_CUHI_0,
    'Arr_CUHI_0_JJA': Arr_CUHI_0_JJA,
    'Arr_CUHI_0_DJF': Arr_CUHI_0_DJF,
    'Arr_SUHI_0': Arr_SUHI_0,
    'Arr_SUHI_0_JJA': Arr_SUHI_0_JJA,
    'Arr_SUHI_0_DJF': Arr_SUHI_0_DJF,
    'Arr_SWBGT_U_0': Arr_SWBGT_U_0,
    'Arr_SWBGT_U_0_JJA': Arr_SWBGT_U_0_JJA,
    'Arr_SWBGT_U_0_DJF': Arr_SWBGT_U_0_DJF,
    'Arr_TBUILD_0': Arr_TBUILD_0,
    'Arr_TBUILD_0_JJA': Arr_TBUILD_0_JJA,
    'Arr_TBUILD_0_DJF': Arr_TBUILD_0_DJF}
df0 = pd.DataFrame(data_dict0)
df0.to_csv('export_hete_temperature0.csv', index=False)
```
