# UHS and TBUILD figure
- This script is used to visualize UHS indice and TBUILD;
- Simulations: CNTL, ROOF_0.9, ROOF_DA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
import textwrap
```

```python
file_path0 = 'exported_data_UHS0.csv'
df0 = pd.read_csv(file_path0)
for column_name in df0.columns:
    globals()[column_name] = df0[column_name]   

file_path1 = 'exported_data_UHS1.csv'
df1 = pd.read_csv(file_path1)
for column_name in df1.columns:
    globals()[column_name] = df1[column_name]  

file_path2 = 'exported_data_UHS2.csv'
df2 = pd.read_csv(file_path2)
for column_name in df2.columns:
    globals()[column_name] = df2[column_name]  
```

```python
# drawing
start_year = 2015
end_year = 2099
linewidth = 0.5
linewidth2 = 1
fontsize = 6
labelfont = 6
labelcolor = '#6b6b6b'
lc1 = '#6e1913'
lc2 = '#a42018'
lc3 = '#f62d18'
lc4 = '#d97c26'
lc5 = '#ff912b'
lc6 = '#30b5d6'
lc7 = '#28ac92'
lc8 = '#1235b6'
lc9 = '#05101f'
lc10 = '#952cc6'
minor_locator = MultipleLocator(10)
specific_years = [2015, 2050, 2099]
alpha=0.075
handletextpad = 0.5
pad = 1
padt = 0.5
length = 4
distance = 5
padding = 4
data = [[[Arr_HIA_U_0, Arr_HIA_U_1,Arr_HIA_U_2], [Arr_HIA_U_0_JJA, Arr_HIA_U_1_JJA, Arr_HIA_U_2_JJA], [Arr_HIA_U_0_DJF, Arr_HIA_U_1_DJF,Arr_HIA_U_2_DJF]],
        [[Arr_APPAR_TEMP_U_0, Arr_APPAR_TEMP_U_1,Arr_APPAR_TEMP_U_2], [Arr_APPAR_TEMP_U_0_JJA, Arr_APPAR_TEMP_U_1_JJA, Arr_APPAR_TEMP_U_2_JJA], [Arr_APPAR_TEMP_U_0_DJF, Arr_APPAR_TEMP_U_1_DJF,Arr_APPAR_TEMP_U_2_DJF]],
        [[Arr_SWBGT_U_0, Arr_SWBGT_U_1,Arr_SWBGT_U_2], [Arr_SWBGT_U_0_JJA, Arr_SWBGT_U_1_JJA, Arr_SWBGT_U_2_JJA], [Arr_SWBGT_U_0_DJF, Arr_SWBGT_U_1_DJF,Arr_SWBGT_U_2_DJF]],
        [[Arr_HUMIDEX_U_0, Arr_HUMIDEX_U_1,Arr_HUMIDEX_U_2], [Arr_HUMIDEX_U_0_JJA, Arr_HUMIDEX_U_1_JJA, Arr_HUMIDEX_U_2_JJA], [Arr_HUMIDEX_U_0_DJF, Arr_HUMIDEX_U_1_DJF,Arr_HUMIDEX_U_2_DJF]],
        [[Arr_DISCOIS_U_0, Arr_DISCOIS_U_1,Arr_DISCOIS_U_2], [Arr_DISCOIS_U_0_JJA, Arr_DISCOIS_U_1_JJA, Arr_DISCOIS_U_2_JJA], [Arr_DISCOIS_U_0_DJF, Arr_DISCOIS_U_1_DJF,Arr_DISCOIS_U_2_DJF]],
        [[Arr_TBUILD_0_C, Arr_TBUILD_1_C,Arr_TBUILD_2_C], [Arr_TBUILD_0_JJA_C, Arr_TBUILD_1_JJA_C, Arr_TBUILD_2_JJA_C], [Arr_TBUILD_0_DJF_C, Arr_TBUILD_1_DJF_C,Arr_TBUILD_2_DJF_C]]
       ]
label = [['CNTL - ANN', 'ROOF_0.9 - ANN', 'ROOF_DA - ANN'], 
         ['CNTL - JJA', 'ROOF_0.9 - JJA', 'ROOF_DA - JJA'], 
         ['CNTL - DJF', 'ROOF_0.9 - DJF', 'ROOF_DA - DJF']]
lc = [lc1, lc3, lc7]
linestyle = ['-', ':', '--']
title = ['(a) NWS_HI', '(b) AT', '(c) SWBGT', '(d) HUMIDEX', '(e) DI', '(f) TBUILD']
la = ['℃', '℃', '℃','℃','','℃']
ymin = 10
ymax = 36
xwidth = 0.75
sizes = 2
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(5.5, 2.75))
ax = axes.flatten()
for i in range(6):
    for j in range(3):
        for m in range(3):
            ax[i].plot(time, data[i][j][m], linewidth=linewidth, linestyle = linestyle[j], label = label[j][m], color = lc[m])
    ax[i].set_title(title[i],fontsize=fontsize,color=labelcolor, pad = padding)   
    ax[i].set_xlim(start_year,end_year)
    ax[i].yaxis.offsetText.set_fontsize(labelfont)
    ax[i].tick_params(axis='y', which='both', left=False, right=False, labelleft=True, labelcolor=labelcolor, labelsize=labelfont, pad=pad, color = labelcolor, width = xwidth)
    ax[i].tick_params(axis='x', which='major', top=False, bottom=True, labelbottom=True, labelcolor=labelcolor, labelsize=labelfont, pad=pad, color = labelcolor, width = xwidth, length = length)
    ax[i].tick_params(axis='x', which='minor', top=False, bottom=True, labelbottom=True, labelcolor=labelcolor, labelsize=labelfont, pad=pad, color = labelcolor, width = xwidth, length = length/2)
    ax[i].spines['left'].set_visible(False)
    ax[i].spines['top'].set_visible(False)
    ax[i].spines['right'].set_visible(False)
    ax[i].spines['bottom'].set_color(labelcolor)
    ax[i].xaxis.set_minor_locator(minor_locator)
    ax[i].set_xticks(specific_years)
    ax[i].grid(True, which='major', axis='y',linestyle='--',linewidth=0.5, alpha=0.5)
    if i == 4:
            ymin = 10
            ymax = 27       
    if i == 5:
            ymin = 17
            ymax = 31
    ax[i].set_ylim(ymin, ymax)
    ax[i].fill_betweenx([ymin, ymax],2021, 2040, color=lc9, alpha=alpha, label='Near-term',edgecolor='none')
    ax[i].fill_betweenx([ymin, ymax],2041, 2060, color=lc5, alpha=alpha, label='Mid-term',edgecolor='none')
    ax[i].fill_betweenx([ymin, ymax],2081, 2100, color=lc2, alpha=alpha, label='Long-term',edgecolor='none')
    ax[i].set_ylabel(la[i], labelpad=5, rotation=0, ha='center', fontsize=labelfont,color=labelcolor)
    ax[i].yaxis.set_label_coords(0, 1.05)

axes[0,0].hlines(y=27, xmin=min(time), xmax=max(time), linewidth=linewidth, color=labelcolor, linestyle='--')
axes[0,0].text(max(time), 27, '27℃', verticalalignment='center', fontsize=labelfont, color=labelcolor)
axes[0,0].scatter(2043, 27, color=lc1, s=sizes)
axes[0,0].scatter(2043, 27, color=lc3, s=sizes)
axes[0,0].scatter(2050, 27, color=lc7, s=sizes)

axes[0,0].hlines(y=33, xmin=min(time), xmax=max(time), linewidth=linewidth, color=labelcolor, linestyle='--')
wrapped_text = textwrap.fill('33℃', width=8)
axes[0,0].text(max(time), 33, wrapped_text, verticalalignment='center', fontsize=labelfont, color=labelcolor)

axes[0,2].hlines(y=29.4, xmin=min(time), xmax=max(time), linewidth=linewidth, color=labelcolor, linestyle='--')
axes[0,2].text(max(time), 29.4, '29.4℃', verticalalignment='center', fontsize=labelfont, color=labelcolor)

axes[1,0].hlines(y=30, xmin=min(time), xmax=max(time), linewidth=linewidth, color=labelcolor, linestyle='--')
wrapped_text2 = textwrap.fill('30℃', width=12)  
axes[1,0].text(max(time), 30, wrapped_text2, verticalalignment='center', fontsize=labelfont, color=labelcolor)
axes[1,0].scatter(2030, 30, color=lc1, s=sizes)
axes[1,0].scatter(2037, 30, color=lc3, s=sizes)
axes[1,0].scatter(2040, 30, color=lc7, s=sizes)

axes[1,1].hlines(y=21, xmin=min(time), xmax=max(time), linewidth=linewidth, color=labelcolor, linestyle='--')
wrapped_text3 = textwrap.fill('<50%', width=12)  
axes[1,1].text(max(time), 21, wrapped_text3, verticalalignment='center', fontsize=labelfont, color=labelcolor)

axes[1,1].hlines(y=24, xmin=min(time), xmax=max(time), linewidth=linewidth, color=labelcolor, linestyle='--')
wrapped_text4 = textwrap.fill('>50%', width=12) 
axes[1,1].text(max(time), 24, wrapped_text4, verticalalignment='center', fontsize=labelfont, color=labelcolor)
axes[1,1].scatter(2080, 24, color=lc1, s=sizes)
axes[1,1].scatter(2084, 24, color=lc3, s=sizes)
axes[1,1].scatter(2084, 24, color=lc7, s=sizes)

axes[1,2].hlines(y=28, xmin=min(time), xmax=max(time), linewidth=linewidth, color=labelcolor, linestyle='--')
wrapped_text2 = textwrap.fill('28℃', width=12)  
axes[1,2].text(max(time), 28, wrapped_text2, verticalalignment='center', fontsize=labelfont, color=labelcolor)
axes[1,2].scatter(2065, 28, color=lc1, s=sizes)
axes[1,2].scatter(2065, 28.1, color=lc3, s=sizes)
axes[1,2].scatter(2094, 28, color=lc7, s=sizes)
axes[1,2].hlines(y=19, xmin=min(time), xmax=max(time), linewidth=linewidth, color=labelcolor, linestyle='--')
wrapped_text2 = textwrap.fill('19℃', width=12) 
axes[1,2].text(max(time), 19, wrapped_text2, verticalalignment='center', fontsize=labelfont, color=labelcolor)    
axes[1,2].scatter(2015, 19, color=lc1, s=sizes)
axes[1,2].scatter(2078, 19, color=lc3, s=sizes)
axes[1,2].scatter(2015, 19, color=lc7, s=sizes)      

lines, labels = axes[0,0].get_legend_handles_labels()

plt.subplots_adjust(left=0.05, right=0.945, bottom=0.225, top=0.93, wspace=0.45, hspace=0.55)
legend = fig.legend(lines, labels, bbox_to_anchor=(0.5, -0.01), loc='lower center', frameon=False, handletextpad=handletextpad, fontsize=labelfont, ncol=4)
lc = [lc1, lc3, lc7, lc1, lc3, lc7, lc1, lc3, lc7, lc9, lc5, lc2]
for i, text in enumerate(legend.get_texts()):
    text.set_color(lc[i])
        
plt.show()
fig.savefig('UHS.png', dpi=300)
```
