```python
import os
import xarray as xr
import numpy as np
import pandas as pd
```

```python
home_path = '/work/n02/n02/yuansun/cesm/'
# surface
sfile = home_path + 'cesm_inputdata/lnd/clm2/surfdata_map/release-clm5.0.18/surfdata_0.9x1.25_hist_16pfts_Irrig_CMIP6_simyr1850_c190214.nc'
#case0
path_0 = home_path + 'archive/case0/lnd/hist/'
#case1
path_1 = home_path + 'archive/case1/lnd/hist/'
#case2
path_2 = home_path + 'archive/case2/lnd/hist/'
```

```python
start_year = 2015
end_year = 2100
```

```python
# case0
Arr_HIA_U_0 = []
Arr_APPAR_TEMP_U_0 = []
Arr_SWBGT_U_0 = []
Arr_HUMIDEX_U_0 = []
Arr_DISCOIS_U_0 = []
Arr_TBUILD_0 = []

Arr_HIA_U_0_JJA = []
Arr_APPAR_TEMP_U_0_JJA = []
Arr_SWBGT_U_0_JJA = []
Arr_HUMIDEX_U_0_JJA = []
Arr_DISCOIS_U_0_JJA = []
Arr_TBUILD_0_JJA = []

Arr_HIA_U_0_DJF = []
Arr_APPAR_TEMP_U_0_DJF = []
Arr_SWBGT_U_0_DJF = []
Arr_HUMIDEX_U_0_DJF = []
Arr_DISCOIS_U_0_DJF = []
Arr_TBUILD_0_DJF = []

for year in range(start_year,end_year):
        fn = 'case0.clm2.h1.' + '%04.0f' % year + '-02-01-00000.nc'
        year_1 = year +1
        fn_1 = 'case0.clm2.h1.' + '%04.0f' % year_1 + '-02-01-00000.nc'
        if(os.path.exists(path_0+fn)):
            ds = xr.open_dataset(path_0+fn)
            ds_1 = xr.open_dataset(path_0+fn_1)
            hia = ds['HIA_U']
            at = ds['APPAR_TEMP_U']
            swbgt = ds['SWBGT_U']
            hi = ds['HUMIDEX_U']
            di = ds['DISCOIS_U']
            tbuild = ds['TBUILD']
            hia_1 = ds_1['HIA_U']
            at_1 = ds_1['APPAR_TEMP_U']
            swbgt_1 = ds_1['SWBGT_U']
            hi_1 = ds_1['HUMIDEX_U']
            di_1 = ds_1['DISCOIS_U']
            tbuild_1 = ds_1['TBUILD']
            
            Arr_HIA_U_0.append(hia.mean().item())
            Arr_APPAR_TEMP_U_0.append(at.mean().item())
            Arr_SWBGT_U_0.append(swbgt.mean().item())
            Arr_HUMIDEX_U_0.append(hi.mean().item())
            Arr_DISCOIS_U_0.append(di.mean().item())
            Arr_TBUILD_0.append(tbuild.mean().item())
            
            Arr_HIA_U_0_JJA.append(hia[6:9,:,:].mean().item())
            Arr_APPAR_TEMP_U_0_JJA.append(at[6:9,:,:].mean().item())
            Arr_SWBGT_U_0_JJA.append(swbgt[6:9,:,:].mean().item())
            Arr_HUMIDEX_U_0_JJA.append(hi[6:9,:,:].mean().item())
            Arr_DISCOIS_U_0_JJA.append(di[6:9,:,:].mean().item())
            Arr_TBUILD_0_JJA.append(tbuild[6:9,:,:].mean().item())
            
            Arr_HIA_U_0_DJF.append(xr.concat([hia.isel(time=11), hia_1.isel(time=slice(0, 2))], dim='time').mean().item())
            Arr_APPAR_TEMP_U_0_DJF.append(xr.concat([at.isel(time=11), at_1.isel(time=slice(0, 2))], dim='time').mean().item())
            Arr_SWBGT_U_0_DJF.append(xr.concat([swbgt.isel(time=11), swbgt_1.isel(time=slice(0, 2))], dim='time').mean().item())
            Arr_HUMIDEX_U_0_DJF.append(xr.concat([hi.isel(time=11), hi_1.isel(time=slice(0, 2))], dim='time').mean().item())
            Arr_DISCOIS_U_0_DJF.append(xr.concat([di.isel(time=11), di_1.isel(time=slice(0, 2))], dim='time').mean().item())
            Arr_TBUILD_0_DJF.append(xr.concat([tbuild.isel(time=11), tbuild_1.isel(time=slice(0, 2))], dim='time').mean().item())
```

```python
Arr_TBUILD_0_C = [value - 273.15 for value in Arr_TBUILD_0]
Arr_TBUILD_0_JJA_C = [value - 273.15 for value in Arr_TBUILD_0_JJA]
Arr_TBUILD_0_DJF_C = [value - 273.15 for value in Arr_TBUILD_0_DJF]
data_dict0 = {
    'Arr_TBUILD_0_C': Arr_TBUILD_0_C,
    'Arr_TBUILD_0_JJA_C': Arr_TBUILD_0_JJA_C,
    'Arr_TBUILD_0_DJF_C': Arr_TBUILD_0_DJF_C,
    'Arr_HIA_U_0': Arr_HIA_U_0,
    'Arr_HIA_U_0_JJA': Arr_HIA_U_0_JJA,
    'Arr_HIA_U_0_DJF': Arr_HIA_U_0_DJF,
    'Arr_APPAR_TEMP_U_0': Arr_APPAR_TEMP_U_0,
    'Arr_APPAR_TEMP_U_0_JJA': Arr_APPAR_TEMP_U_0_JJA,
    'Arr_APPAR_TEMP_U_0_DJF': Arr_APPAR_TEMP_U_0_DJF,
    'Arr_SWBGT_U_0': Arr_SWBGT_U_0,
    'Arr_SWBGT_U_0_JJA': Arr_SWBGT_U_0_JJA,
    'Arr_SWBGT_U_0_DJF': Arr_SWBGT_U_0_DJF,
    'Arr_HUMIDEX_U_0': Arr_HUMIDEX_U_0,
    'Arr_HUMIDEX_U_0_JJA': Arr_HUMIDEX_U_0_JJA,
    'Arr_HUMIDEX_U_0_DJF': Arr_HUMIDEX_U_0_DJF,
    'Arr_DISCOIS_U_0': Arr_DISCOIS_U_0,
    'Arr_DISCOIS_U_0_JJA': Arr_DISCOIS_U_0_JJA,
    'Arr_DISCOIS_U_0_DJF': Arr_DISCOIS_U_0_DJF}
df0 = pd.DataFrame(data_dict0)
df0.to_csv('exported_data_UHS0.csv', index=False)
```
