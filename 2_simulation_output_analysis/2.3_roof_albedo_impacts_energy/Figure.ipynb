# Roof-related heat flux figure
- This script is used to export roof-related heat flux variables;
- Simulations: CNTL, ROOF_0.9, ROOF_DA;

```python
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

```python
file_path0 = 'exported_hete_energy0.csv'
df0 = pd.read_csv(file_path0)
for column_name in df0.columns:
    globals()[column_name] = df0[column_name]  

file_path1 = 'exported_hete_energy1.csv'
df1 = pd.read_csv(file_path1)
for column_name in df1.columns:
    globals()[column_name] = df1[column_name]  

file_path2 = 'exported_hete_energy2.csv'
df2 = pd.read_csv(file_path2)
for column_name in df2.columns:
    globals()[column_name] = df2[column_name]  
```

```python
# drawing
start_year = 2015
end_year = 2099
linewidth = 0.5
linewidth2 = 1
fontsize = 6
labelfont = 6
labelcolor = '#6b6b6b'
lc1 = '#6e1913'
lc2 = '#a42018'
lc3 = '#f62d18'
lc4 = '#d97c26'
lc5 = '#ff912b'
lc6 = '#30b5d6'
lc7 = '#28ac92'
lc8 = '#1235b6'
lc9 = '#05101f'
lc10 = '#952cc6'
minor_locator = MultipleLocator(10)
specific_years = [2015, 2050, 2099]
alpha=0.075
handletextpad = 0.5
pad = 1
padt = 0.5
length = 4
distance = 5
padding = 4
data = [[[Arr_FSA_U_0, Arr_FSA_U_1,Arr_FSA_U_2], 
         [Arr_FSA_U_0_JJA, Arr_FSA_U_1_JJA, Arr_FSA_U_2_JJA], 
         [Arr_FSA_U_0_DJF, Arr_FSA_U_1_DJF,Arr_FSA_U_2_DJF]],
        [[Arr_FIRA_U_0, Arr_FIRA_U_1,Arr_FIRA_U_2], 
         [Arr_FIRA_U_0_JJA, Arr_FIRA_U_1_JJA, Arr_FIRA_U_2_JJA], 
         [Arr_FIRA_U_0_DJF, Arr_FIRA_U_1_DJF,Arr_FIRA_U_2_DJF]],
        [[Arr_FSH_U_0, Arr_FSH_U_1,Arr_FSH_U_2], 
         [Arr_FSH_U_0_JJA, Arr_FSH_U_1_JJA, Arr_FSH_U_2_JJA], 
         [Arr_FSH_U_0_DJF, Arr_FSH_U_1_DJF,Arr_FSH_U_2_DJF]],
        [[Arr_EFLX_LH_TOT_U_0, Arr_EFLX_LH_TOT_U_1,Arr_EFLX_LH_TOT_U_2], 
         [Arr_EFLX_LH_TOT_U_0_JJA, Arr_EFLX_LH_TOT_U_1_JJA, Arr_EFLX_LH_TOT_U_2_JJA], 
         [Arr_EFLX_LH_TOT_U_0_DJF, Arr_EFLX_LH_TOT_U_1_DJF,Arr_EFLX_LH_TOT_U_2_DJF]],
        [[Arr_FGR_U_0, Arr_FGR_U_1,Arr_FGR_U_2], 
         [Arr_FGR_U_0_JJA, Arr_FGR_U_1_JJA, Arr_FGR_U_2_JJA], 
         [Arr_FGR_U_0_DJF, Arr_FGR_U_1_DJF,Arr_FGR_U_2_DJF]],
        [[Arr_AHF_0, Arr_AHF_1,Arr_AHF_2], 
         [Arr_AHF_0_JJA, Arr_AHF_1_JJA, Arr_AHF_2_JJA], 
         [Arr_AHF_0_DJF, Arr_AHF_1_DJF,Arr_AHF_2_DJF]],
        [[Arr_AC_0, Arr_AC_1,Arr_AC_2], 
         [Arr_AC_0_JJA, Arr_AC_1_JJA, Arr_AC_2_JJA], 
         [Arr_AC_0_DJF, Arr_AC_1_DJF,Arr_AC_2_DJF]],
        [[Arr_HEAT_0, Arr_HEAT_1,Arr_HEAT_2], 
         [Arr_HEAT_0_JJA, Arr_HEAT_1_JJA, Arr_HEAT_2_JJA], 
         [Arr_HEAT_0_DJF, Arr_HEAT_1_DJF,Arr_HEAT_2_DJF]],
        [[Arr_WASTEHEAT_0, Arr_WASTEHEAT_1,Arr_WASTEHEAT_2], 
         [Arr_WASTEHEAT_0_JJA, Arr_WASTEHEAT_1_JJA, Arr_WASTEHEAT_2_JJA], 
         [Arr_WASTEHEAT_0_DJF, Arr_WASTEHEAT_1_DJF,Arr_WASTEHEAT_2_DJF]],
       ]
label = [['CNTL - ANN', 'ROOF_0.9 - ANN', 'ROOF_DA - ANN'], 
         ['CNTL - JJA', 'ROOF_0.9 - JJA', 'ROOF_DA - JJA'], 
         ['CNTL - DJF', 'ROOF_0.9 - DJF', 'ROOF_DA - DJF']]
lc = [lc1, lc3, lc7]
linestyle = ['-', ':', '--']
title = ['(a) FSA', '(b) FIRA', '(c) FSH', '(d) FLH', '(e) FGR', '(f) AHF','(g) AC', '(h) HEAT', '(i) WASTEHEAT']
ymin = [0, 0, 0, 0, -15, 0,0, 0, 0]
ymax = [200, 80, 80, 80, 5, 0.3, 0.0175, 0.3, 0.05]
xwidth = 0.75
```

```python
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(5.5, 4))
ax = axes.flatten()
for i in range(9):
    for j in range(3):
        for m in range(3):
            if i ==0:
               ax[i].plot(time, data[i][j][m], linewidth=linewidth, linestyle = linestyle[j], label = label[j][m], color = lc[m])
            else:
               ax[i].plot(time, data[i][j][m], linewidth=linewidth, linestyle = linestyle[j], color = lc[m]) 
    ax[i].set_title(title[i],fontsize=fontsize,color=labelcolor, pad = padding)   
    ax[i].set_xlim(start_year,end_year)
    ax[i].tick_params(axis='y', which='both', left=False, right=False, labelleft=True, labelcolor=labelcolor, labelsize=labelfont, pad=pad, color = labelcolor, width = xwidth)
    ax[i].tick_params(axis='x', which='major', top=False, bottom=True, labelbottom=True, labelcolor=labelcolor, labelsize=labelfont, pad=pad, color = labelcolor, width = xwidth, length = length)
    ax[i].tick_params(axis='x', which='minor', top=False, bottom=True, labelbottom=True, labelcolor=labelcolor, labelsize=labelfont, pad=pad, color = labelcolor, width = xwidth, length = length/2)
    ax[i].spines['left'].set_visible(False)
    ax[i].spines['top'].set_visible(False)
    ax[i].spines['right'].set_visible(False)
    ax[i].spines['bottom'].set_color(labelcolor)
    ax[i].xaxis.set_minor_locator(minor_locator)
    ax[i].set_xticks(specific_years)
    ax[i].grid(True, which='major', axis='y',linestyle='--',linewidth=0.5, alpha=0.5)
    ax[i].set_ylim(ymin[i], ymax[i])
    ax[i].fill_betweenx([ymin[i], ymax[i]],2021, 2040, color=lc9, alpha=alpha, label='Near-term',edgecolor='none')
    ax[i].fill_betweenx([ymin[i], ymax[i]],2041, 2060, color=lc5, alpha=alpha, label='Mid-term',edgecolor='none')
    ax[i].fill_betweenx([ymin[i], ymax[i]],2081, 2100, color=lc2, alpha=alpha, label='Long-term',edgecolor='none')
    ax[i].set_ylabel('W m⁻²', labelpad=5, rotation=0, ha='center', fontsize=labelfont,color=labelcolor)
    ax[i].yaxis.set_label_coords(0.1, 1.05)

plt.subplots_adjust(left=0.065, right=0.97, bottom=0.15, top=0.96, wspace=0.275, hspace= 0.55)
lines, l = ax[0].get_legend_handles_labels()
legend = fig.legend(lines, l, bbox_to_anchor=(0.5, -0.01), loc='lower center',frameon=False,handletextpad=handletextpad,fontsize=labelfont, ncol = 4)

cll = [lc1, lc3, lc7, lc1, lc3, lc7, lc1, lc3, lc7, lc9, lc5, lc2]

for i, text in enumerate(legend.get_texts()):
    text.set_color(cll[i])
    
fig.savefig('roof_energy.png', dpi=300)      
```
