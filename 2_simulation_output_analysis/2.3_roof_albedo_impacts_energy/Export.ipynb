```python
import os
import xarray as xr
import numpy as np
import pandas as pd
```

```python
home_path = '/work/n02/n02/yuansun/'
# surface
sfile = home_path + 'cesm/cesm_inputdata/lnd/clm2/surfdata_map/release-clm5.0.18/surfdata_0.9x1.25_hist_16pfts_Irrig_CMIP6_simyr1850_c190214.nc'
#case0
path_0 = home_path + 'cesm/archive/case0/lnd/hist/'
#case1
path_1 = home_path + 'cesm/archive/case1/lnd/hist/'
#case2
path_2 = home_path + 'cesm/archive/case2/lnd/hist/'
```

```python
start_year = 2015
end_year = 2100
```

```python
default = xr.open_dataset(sfile)
mask = np.any(default['PCT_URBAN'] != 0, axis=0)
mask = mask.rename({'lsmlat': 'lat', 'lsmlon': 'lon'})
```

```python
Arr_FIRA_U_0 = []
Arr_FSA_U_0 = []
Arr_FSH_U_0 = []
Arr_EFLX_LH_TOT_U_0 = []
Arr_FGR_U_0 = []
Arr_AC_0 = []
Arr_HEAT_0 = []
Arr_AHF_0 = []
Arr_WASTEHEAT_0 = []

Arr_FIRA_U_0_JJA = []
Arr_FSA_U_0_JJA = []
Arr_FSH_U_0_JJA = []
Arr_EFLX_LH_TOT_U_0_JJA = []
Arr_FGR_U_0_JJA = []
Arr_AC_0_JJA = []
Arr_HEAT_0_JJA = []
Arr_AHF_0_JJA = []
Arr_WASTEHEAT_0_JJA = []

Arr_FIRA_U_0_DJF = []
Arr_FSA_U_0_DJF = []
Arr_FSH_U_0_DJF = []
Arr_EFLX_LH_TOT_U_0_DJF = []
Arr_FGR_U_0_DJF = []
Arr_AC_0_DJF = []
Arr_HEAT_0_DJF = []
Arr_AHF_0_DJF = []
Arr_WASTEHEAT_0_DJF = []

time = []
for year in range(start_year,end_year):
    fn = 'case0.clm2.h1.' + '%04.0f' % year + '-02-01-00000.nc'
    year_1 = year+1
    fn_1 = 'case0.clm2.h1.' + '%04.0f' % year_1 + '-02-01-00000.nc'
    if(os.path.exists(path_0+fn)):
        ds = xr.open_dataset(path_0+fn)
        ds_1 = xr.open_dataset(path_0+fn_1)
        time.append(year)
        fira= ds['FIRA_U']
        fsa = ds['FSA_U']
        fsh = ds['FSH_U']
        eflx = ds['EFLX_LH_TOT_U']
        fgr = ds['FGR_U']
        ac = ds['URBAN_AC'].where(mask)
        heat = ds['URBAN_HEAT'].where(mask)
        wasteheat = ds['WASTEHEAT'].where(mask)
        ahf = heat + wasteheat
        
        fira_1= ds_1['FIRA_U']
        fsa_1 = ds_1['FSA_U']
        fsh_1 = ds_1['FSH_U']
        eflx_1 = ds_1['EFLX_LH_TOT_U']
        fgr_1 = ds_1['FGR_U']
        ac_1 = ds_1['URBAN_AC'].where(mask)
        heat_1 = ds_1['URBAN_HEAT'].where(mask)
        wasteheat_1 = ds_1['WASTEHEAT'].where(mask)
        ahf_1 = heat_1 + wasteheat_1
        
        Arr_FIRA_U_0.append(fira.mean().item())
        Arr_FSA_U_0.append(fsa.mean().item())
        Arr_FSH_U_0.append(fsh.mean().item())
        Arr_EFLX_LH_TOT_U_0.append(eflx.mean().item())
        Arr_FGR_U_0.append(fgr.mean().item())
        Arr_AC_0.append(ac.mean().item())
        Arr_HEAT_0.append(heat.mean().item())
        Arr_AHF_0.append(ahf.mean().item())
        Arr_WASTEHEAT_0.append(wasteheat.mean().item())
        
        Arr_FIRA_U_0_JJA.append(fira[6:9,:,:].mean().item())
        Arr_FSA_U_0_JJA.append(fsa[6:9,:,:].mean().item())
        Arr_FSH_U_0_JJA.append(fsh[6:9,:,:].mean().item())
        Arr_EFLX_LH_TOT_U_0_JJA.append(eflx[6:9,:,:].mean().item())
        Arr_FGR_U_0_JJA.append(fgr[6:9,:,:].mean().item())
        Arr_AC_0_JJA.append(ac[6:9,:,:].mean().item())
        Arr_HEAT_0_JJA.append(heat[6:9,:,:].mean().item())
        Arr_AHF_0_JJA.append(ahf[6:9,:,:].mean().item())
        Arr_WASTEHEAT_0_JJA.append(wasteheat[6:9,:,:].mean().item())
        
        Arr_FIRA_U_0_DJF.append(xr.concat([fira.isel(time=11), fira_1.isel(time=slice(0, 2))], dim='time').mean().item())
        Arr_FSA_U_0_DJF.append(xr.concat([fsa.isel(time=11), fsa_1.isel(time=slice(0, 2))], dim='time').mean().item())
        Arr_FSH_U_0_DJF.append(xr.concat([fsh.isel(time=11), fsh_1.isel(time=slice(0, 2))], dim='time').mean().item())
        Arr_EFLX_LH_TOT_U_0_DJF.append(xr.concat([eflx.isel(time=11), eflx_1.isel(time=slice(0, 2))], dim='time').mean().item())
        Arr_FGR_U_0_DJF.append(xr.concat([fgr.isel(time=11), fgr_1.isel(time=slice(0, 2))], dim='time').mean().item())
        Arr_AC_0_DJF.append(xr.concat([ac.isel(time=11), ac.isel(time=slice(0, 2))], dim='time').mean().item())
        Arr_HEAT_0_DJF.append(xr.concat([heat.isel(time=11), heat_1.isel(time=slice(0, 2))], dim='time').mean().item())
        Arr_AHF_0_DJF.append(xr.concat([ahf.isel(time=11), ahf_1.isel(time=slice(0, 2))], dim='time').mean().item())
        Arr_WASTEHEAT_0_DJF.append(xr.concat([wasteheat.isel(time=11), wasteheat_1.isel(time=slice(0, 2))], dim='time').mean().item())
```
