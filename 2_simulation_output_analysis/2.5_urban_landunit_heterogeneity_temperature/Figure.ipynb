# Landunit level figure
- This script is used to export CUHI, SUHI, SWBGT, TBUILD variables at the land unit level;
- Simulations: CNTL, ROOF_DA, IMPRAOD_DA, WALL_DA

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
```

```python
f0 = 'mon_0.csv'
df0 = pd.read_csv(f0)
f2 = 'mon_2.csv'
df2 = pd.read_csv(f2)
f3 = 'mon_3.csv'
df3 = pd.read_csv(f3)
f5 = 'mon_5.csv'
df5 = pd.read_csv(f5)

for column_name in df0.columns:
    globals()[column_name] = df0[column_name]

for column_name in df2.columns:
    globals()[column_name] = df2[column_name] 
    
for column_name in df3.columns:
    globals()[column_name] = df3[column_name]      

for column_name in df5.columns:
    globals()[column_name] = df5[column_name]         
```

```python
day_tbd = [UHImax_tbd_0, UHImax_tbd_2, UHImax_tbd_3, UHImax_tbd_5]
day_hd = [UHImax_hd_0, UHImax_hd_2, UHImax_hd_3, UHImax_hd_5]
day_md = [UHImax_md_0, UHImax_md_2, UHImax_md_3, UHImax_md_5]

night_tbd = [UHImin_tbd_0, UHImin_tbd_2, UHImin_tbd_3, UHImin_tbd_5]
night_hd = [UHImin_hd_0, UHImin_hd_2, UHImin_hd_3, UHImin_hd_5]
night_md = [UHImin_md_0, UHImin_md_2, UHImin_md_3, UHImin_md_5]

swbgt_tbd = [SWBGT_U_tbd_0, SWBGT_U_tbd_2, SWBGT_U_tbd_3, SWBGT_U_tbd_5]
swbgt_hd = [SWBGT_U_hd_0, SWBGT_U_hd_2, SWBGT_U_hd_3, SWBGT_U_hd_5]
swbgt_md = [SWBGT_U_md_0, SWBGT_U_md_2, SWBGT_U_md_3, SWBGT_U_md_5]

tbuild_tbd = [TBUILD_tbd_0, TBUILD_tbd_2, TBUILD_tbd_3, TBUILD_tbd_5]
tbuild_hd = [TBUILD_hd_0, TBUILD_hd_2, TBUILD_hd_3, TBUILD_hd_5]
tbuild_md = [TBUILD_md_0, TBUILD_md_2, TBUILD_md_3, TBUILD_md_5]
```

```python
data = [[day_tbd, day_hd, day_md], 
        [night_tbd, night_hd, night_md], 
        [swbgt_tbd, swbgt_hd, swbgt_md], 
        [tbuild_tbd, tbuild_hd, tbuild_md]]
```

```python
delta_data = [[[UHImax_tbd_2-UHImax_tbd_0, UHImax_tbd_3-UHImax_tbd_0, UHImax_tbd_5-UHImax_tbd_0],
               [UHImax_hd_2-UHImax_hd_0, UHImax_hd_3-UHImax_hd_0, UHImax_hd_5-UHImax_hd_0],
               [UHImax_md_2-UHImax_md_0, UHImax_md_3-UHImax_md_0, UHImax_md_5-UHImax_md_0]],
             [[UHImin_tbd_2-UHImin_tbd_0, UHImin_tbd_3-UHImin_tbd_0, UHImin_tbd_5-UHImin_tbd_0],
              [UHImin_hd_2-UHImin_hd_0, UHImin_hd_3-UHImin_hd_0, UHImin_hd_5-UHImin_hd_0],
              [UHImin_md_2-UHImin_md_0, UHImin_md_3-UHImin_md_0, UHImin_md_5-UHImin_md_0]],
             [[SWBGT_U_tbd_2-SWBGT_U_tbd_0, SWBGT_U_tbd_3-SWBGT_U_tbd_0, SWBGT_U_tbd_5-SWBGT_U_tbd_0],
              [SWBGT_U_hd_2-SWBGT_U_hd_0, SWBGT_U_hd_3-SWBGT_U_hd_0, SWBGT_U_hd_5-SWBGT_U_hd_0],
              [SWBGT_U_md_2-SWBGT_U_md_0, SWBGT_U_md_3-SWBGT_U_md_0, SWBGT_U_md_5-SWBGT_U_md_0]],
             [[TBUILD_tbd_2-TBUILD_tbd_0, TBUILD_tbd_3-TBUILD_tbd_0,TBUILD_tbd_5-TBUILD_tbd_0],
              [TBUILD_hd_2-TBUILD_hd_0, TBUILD_hd_3-TBUILD_hd_0,TBUILD_hd_5-TBUILD_hd_0],
              [TBUILD_md_2-TBUILD_md_0, TBUILD_md_3-TBUILD_md_0,TBUILD_md_5-TBUILD_md_0]]]
```

```python
# drawing
linewidth = 0.5
linewidth2 = 1
fontsize = 6
start_year = 2015
end_year = 2100
labelcolor = '#6b6b6b'
lc1 = '#6e1913'
lc2 = '#a42018'
lc3 = '#f62d18'
lc4 = '#d97c26'
lc5 = '#ff912b'
lc6 = '#30b5d6'
lc7 = '#28ac92'
lc8 = '#123264'
lc9 = '#05101f'
lc10 = '#952cc6'
lc11 = '#2c4dc6'
lc12 = '#5fc62c'
minor_locator = MultipleLocator(10)
specific_years = [2015, 2050, 2099]
alpha=0.075
xwidth = 0.75
length = 4
labelfont = 6
linestyle='dashed'
pad=1
handletextpad=0.5
l = ['CNTL', 'ROOF_DA', 'IMPROAD_DA', 'WALL_DA']
lc = [lc1, lc3, lc7, lc4, lc9, lc5, lc2]
t = [['(a) DAY - TBD','(b) DAY - HD','(c) DAY - MD'],
     ['(d) NIGHT - TBD', '(e) NIGHT - HD', '(f) NIGHT - MD'],
     ['(g) SWBGT - TBD', '(h) SWBGT - HD', '(i) SWBGT - MD'],
     ['(j) TBUILD - TBD', '(k) TBUILD - HD', '(l) TBUILD - MD']]

ymin1 = -0.5
ymax1 = 3
ymin3 = 20
ymax3 = 30
ymin_list = [ymin1, ymin1, ymin3, ymin3]
ymax_list = [ymax1, ymax1,ymax3,ymax3]
time = list(range(2015, 2100))
y1 = 0.5
y2 = 1.2
delta_ymax = [[y1, y1,y1],[1, y1, y1], [y2,y2, y2,y2], [y2, y2,y2,y2]]
```

```python
fig, axes = plt.subplots(nrows=4, ncols=3, figsize=(5.5, 4))

for i in range(4):
    ymin, ymax = ymin_list[i], ymax_list[i]
    for j in range(3):
        for m in range(4):
            axes[i,j].plot(time, data[i][j][m], linewidth=linewidth,label=l[m], color = lc[m])
            if m==3:
               axes[i,j].fill_betweenx([ymin, ymax],2021, 2040, color=lc9, alpha=alpha, label='Near-term', edgecolor='none')
               axes[i,j].fill_betweenx([ymin, ymax],2041, 2060, color=lc5, alpha=alpha, label='Mid-term', edgecolor='none')
               axes[i,j].fill_betweenx([ymin, ymax],2081, 2100, color=lc2, alpha=alpha, label='Long-term', edgecolor='none')
               axes[i,j].text(2015,ymax-delta_ymax[i][j],': ' + str("{:.2f}".format(delta_data[i][j][m-3][84]))+' ', fontsize=labelfont, color=lc[m-2])
               axes[i,j].text(2041,ymax-delta_ymax[i][j],': ' + str("{:.2f}".format(delta_data[i][j][m-2][84]))+' ', fontsize=labelfont, color=lc[m-1]) 
               axes[i,j].text(2067,ymax-delta_ymax[i][j],': ' + str("{:.2f}".format(delta_data[i][j][m-1][84]))+'', fontsize=labelfont, color=lc[m]) 
        axes[i,j].set_xlim(start_year,2099)
        axes[i,j].tick_params(axis='y', which='major', left=False, right=False, labelleft=True, labelcolor=labelcolor,labelsize=labelfont, color = labelcolor, pad = pad, width = xwidth, length = 0)
        axes[i,j].tick_params(axis='x', which='minor', top=False, bottom=True, labelbottom=True, labelcolor=labelcolor,labelsize=labelfont, color = labelcolor,  pad = pad, width = xwidth, length = length/2)
        axes[i,j].tick_params(axis='x', which='major', top=False, bottom=True, labelbottom=True, labelcolor=labelcolor,labelsize=labelfont, color = labelcolor, pad = pad, width = xwidth, length = length)
        axes[i,j].spines['left'].set_visible(False)
        axes[i,j].spines['top'].set_visible(False)
        axes[i,j].spines['right'].set_visible(False)
        axes[i,j].spines['bottom'].set_color(labelcolor)
        axes[i,j].spines['bottom'].set_color(labelcolor)
        axes[i,j].xaxis.set_minor_locator(minor_locator)
        axes[i,j].set_xticks(specific_years)
        axes[i,j].grid(True, which='major', axis='y',linestyle='--',linewidth=linewidth, alpha=0.5)
        
        axes[i,j].set_ylim(ymin, ymax)
        
        axes[i,j].set_title(t[i][j],fontsize=fontsize,color=labelcolor, pad = 3)
        axes[i,j].set_ylabel('', labelpad=5, rotation=0, ha='center', fontsize=labelfont,color=labelcolor)
        axes[i,j].yaxis.set_label_coords(0, 1.06)
        
lines, labels = axes[0,0].get_legend_handles_labels()
new_order = [ 0, 4, 1, 5, 2, 6, 3]
new_lines = [lines[i] for i in new_order]
new_labels = [labels[i] for i in new_order]
new_lc = [lc[i] for i in new_order]
legend = fig.legend(new_lines, new_labels, bbox_to_anchor=(0.5, -0.01), loc='lower center',frameon=False,handletextpad=handletextpad,fontsize=labelfont,ncol=4)

for i, text in enumerate(legend.get_texts()):
    text.set_color(new_lc[i])
    
plt.subplots_adjust(left=0.037, right=0.97, bottom=0.12, top=0.96, wspace=0.22, hspace= 0.55)        
plt.show()
fig.savefig('hete_numurbl.png', dpi=300)
```

## Calculate slope

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import statsmodels.api as sm
```

```python
rf = pd.read_csv('../hete_temperature/alb_dynroof_3.csv')
for column_name in rf.columns:
    globals()[column_name] = rf[column_name]  

wl = pd.read_csv('../hete_temperature/alb_dynwall_3.csv')
for column_name in wl.columns:
    globals()[column_name] = wl[column_name]

rd = pd.read_csv('../hete_temperature/alb_dynrd_3.csv')
for column_name in rd.columns:
    globals()[column_name] = rd[column_name]
```

```python
rf
```

```python
df = pd.DataFrame(UHImax_tbd_2)
x = rf[rf['numurbl']==0].reset_index()[['alb']]
y = df['UHImax_tbd_2']
model = LinearRegression()
model.fit(x, y)
slope = model.coef_[0]  
intercept = model.intercept_  
y_pred = model.predict(x)
r2 = r2_score(y, y_pred)

x_with_intercept = sm.add_constant(x)
model_stats = sm.OLS(y, x_with_intercept).fit()
print(model_stats.summary())
print(f"Slope (Coefficient for 'time'): {slope}")
print(f"Intercept: {intercept}")
print(f"R-squared: {r2}")
```
